# [Basic Concepts of NLP](https://towardsdatascience.com/basic-concepts-of-natural-language-processing-nlp-models-and-python-implementation-88a589ce1fc0/)

```python
line = 'Reaching out for HELP. Please meet me in LONDON at 6 a.m xyz@abc.com #urgent'
```

```python
# 1. Remove stopword
import nltk
import re

nltk.download("stopwords", quiet=True)

extra_list = ["let", "may", "might", "must", "need", "apologies", "meet"]
stopword = nltk.corpus.stopwords.words("english")
stopword.extend(extra_list)

line = ' '.join([i for i in line.split() if i not in stopword])
```

```python
# 2. Make lower case
line = line.lower()
```

```python
# 3. Lemmatization
nltk.download("wordnet", quiet=True)

def lemmatize_text(text):
  w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
  lemmatizer = nltk.stem.WordNetLemmatizer()

  return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]

lemmatize_text(line)
```

```python
# 4. Stemming:
def stem_porter(text):
  w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
  ps = nltk.PorterStemmer()
  return [ps.stem(w) for w in w_tokenizer.tokenize(text)]

stem_porter(line)
```

```python
# 5. Removal of regular expression
line = re.sub('S*@S*s?'," ",line) #email remove
line = re.sub('s+'," ",line) #new line character remove
line = re.sub("'"," ",line) #single quote remove
line = re.sub('_'," ",line) #underscore remove
line = re.sub('httpS*s?'," ",line) #link remove
line = ' '.join([i for i in line.split() if i.find('#') < 0])
line = ' '.join([i for i in line.split() if i in re.findall(r'w+',line)])
```

```python
# 6. Parts-of-Speech (POS) tagging
import spacy
from spacy.tokenizer import Tokenizer

nlp = spacy.load("en_core_web_sm")
tokens_spacy = nlp(line)

for token in tokens_spacy:
  print(token.text, ': ', token.pos_, ': ', token.is_stop)
```
