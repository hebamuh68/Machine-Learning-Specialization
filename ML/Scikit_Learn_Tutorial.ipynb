# [**Data Collection**](https://www.datacamp.com/tutorial/machine-learning-python)

```python
import pandas as pd
from sklearn.datasets import load_wine

wine_data = load_wine()
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
wine_df['target'] = wine_data.target

wine_df.head()
```



---


# **Data Exploration**

```python
# Summary of the wine data contained within the DataFrame
wine_df.info()
```

```python
# Descriptive statistics about each feature in the dataset.
wine_df.describe()
```



---


# **Data preprocessing**

```python
from sklearn.preprocessing import StandardScaler

# Split data into features and label
wine_cop = wine_df[wine_data.feature_names].copy()
labels_cop = wine_df['target'].copy()

# Instantiate scaler and fit on features
scaler = StandardScaler()
scaler.fit(wine_cop)

# Transform features
wine_cop_scaled = scaler.transform(wine_cop.values)
wine_cop_scaled[0:3]
```

# **Model training**

```python
from sklearn.model_selection import  train_test_split

wine_cop_scaled_train, wine_cop_scaled_test, labels_cop_train, labels_cop_test = train_test_split(wine_cop_scaled,
                                                        labels_cop,
                                                        train_size=.7,
                                                        random_state=25)
```

```python
# Building the model
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

log_reg = LogisticRegression()
svm = SVC()
tree = DecisionTreeClassifier()

log_reg.fit(wine_cop_scaled_train, labels_cop_train)
svm.fit(wine_cop_scaled_train, labels_cop_train)
tree.fit(wine_cop_scaled_train, labels_cop_train)

log_reg_preds = log_reg.predict(wine_cop_scaled_test)
svm_preds = svm.predict(wine_cop_scaled_test)
tree_preds = tree.predict(wine_cop_scaled_test)
```



---


# Model evaluation

```python
from sklearn.metrics import classification_report

model_preds = {
    "Logistic Regression": log_reg_preds,
    "Support Vector Machine": svm_preds,
    "Decision Tree": tree_preds
}

for model, preds in model_preds.items():
  print(f"{model} Results:\n{classification_report(labels_cop_test, preds)}", sep="\n\n")
```

# **Decoding the Report**

*   **Precision:** "Of all the times the model guessed 'Wine Class 0', how many were actually correct?" (High precision = low "false alarms").

*   **Recall:** "Of all the actual 'Wine Class 0' samples in the data, how many did the model find?" (High recall = the model didn't miss many).

*   **F1-Score:** A balance between Precision and Recall. It’s the best "all-around" metric if you can't decide which is more important.

*   **Support:** Simply the number of samples for that class in your test set (e.g., there were 17 samples of Class 0).
